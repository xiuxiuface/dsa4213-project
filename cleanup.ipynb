```python
import nbformat

# Load the existing notebook (read-only)
nb = nbformat.read("RAG_QA.ipynb", as_version=4)

# Remove widget/metadata junk
for cell in nb.cells:
    if "metadata" in cell:
        cell["metadata"].pop("widgets", None)
        cell["metadata"].pop("state", None)

# Also clean notebook-level metadata
nb["metadata"].pop("widgets", None)
nb["metadata"].pop("state", None)

# Save a clean copy
nbformat.write(nb, "RAG_QA_clean.ipynb")

print("✅ Clean notebook saved as RAG_Summarization_clean.ipynb")
```
